```python
import matplotlib.pyplot as plt

sns.scatterplot(
    data=df,
    x='gdp_per_capita',
    y='depressive_disorder_rate',
    hue='country',  # this ensures same region = same color
    legend = False

)

plt.title("Depression Rate vs GDP per Capita by Region")
plt.xlabel("gdp_per_capita")
plt.ylabel("depressive_disorder_rate(%)")
plt.legend(title='country')
plt.savefig("all_countries_depression_rate.png")
plt.show()
```
